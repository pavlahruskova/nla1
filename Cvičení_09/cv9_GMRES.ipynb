## GMRES – Zobecněná metoda minimálních reziduí

Budeme řešit soustavu: $ Ax = b $

Metoda GMRES hledá řešení v Krylovově podprostoru:
$$
\mathcal{K}^k(A, r_0) = \langle r_0, Ar_0, A^2 r_0, \dots \rangle . 
$$

Pro obecnou matici $A$ metoda GMRES nevytváří $A$-ortogonální bázi, ale pouze eukleidovsky ortonormální bázi. 


```python

import numpy as np
import numpy.linalg as la
```


## Arnoldiho algoritmus

je speciální varianta Gramm-Schmidtova algoritmu. GMRES využívá Arnoldiho algoritmus pro konstrukci ortonormální báze Krylovova prostoru.

Výstupem je:
- ortonormální báze $V$
- Hessenbergova matice $H$

Doplňte implementaci Arnoldiho procesu.


Dle algoritmu z přednášek: 

$$  
\begin{array}{l}
    r^{0} := b - Ax^k \\
    \ \\
    r^{0} := \frac{r^0} {\| r^0 \|}\\ 
    \ \\
    v^{k+1} := Ar^k - \sum_{l=0}^k \beta_{kl} r^l,
    \text{ kde } \ \beta_{kl} := (v^k)^T A^T v^l\\ 
    \ \\
    v^{k+1} := \frac{r^{k+1}} {\| r^{k+1} \|}\\ 
    \ \\
\end{array}\\
$$

Jednotlivé iterace GMRES hledáme ve tvaru: 
$$  
    x^{k+1} := x^0 + \sum_{l=0}^k \alpha_{l} v^l.
$$

```python

def arnoldi(A, v0, k):
    n = A.shape[0]
    V = np.zeros((n, k+1))
    H = np.zeros((k+1, k))

    V[:, 0] = v0 / la.norm(v0)

    for j in range(k):
        w = A @ V[:, j]

        # ortogonalizace
        for i in range(j+1):
            H[i, j] = np.dot(V[:, i], w)
            w = w - H[i, j] * V[:, i]

        H[j+1, j] = la.norm(w)

        if H[j+1, j] == 0:
            break

        V[:, j+1] = w / H[j+1, j]

    return V, H
```
